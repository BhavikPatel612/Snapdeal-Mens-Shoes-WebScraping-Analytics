```python
import requests
from bs4 import BeautifulSoup as dff
```

```python
url="https://www.snapdeal.com/products/mens-footwear-sports-shoes?sort=plrty&q=Brand%3AAbros%5EAIRCON%5EAirhype%7C#bcrumbSearch:shoes%20for%20men|bcrumbLabelId:46138273"
headers= {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"}
response=requests.get(url,headers=headers)
response
```

```python
soup=dff(response.text,"html.parser")
soup
```

```python
name=soup.find_all("p",class_="product-title")
brand=[]
for i in name:
    j=i.text.strip().split()[0]
    brand.append(j)
print((brand))    
```

```python
product1=soup.find_all("p",class_="product-title")
product_name=[]
for i in product1:
    j=" ".join(i.text.strip().split()[1:])
    product_name.append(j)
print(product_name)
```

```python
original=soup.find_all("span",class_="lfloat product-desc-price strike")
original_price=[]
for i in original:
    j=i.text.replace("Rs.","").strip().replace(",","")
    original_price.append(j)
print(original_price)
```

```python
original_price=soup.find_all("span",class_="lfloat product-price")
discount_price=[]
for i in original_price:
    j=i.text.replace("Rs.","").replace(",","").strip()
    discount_price.append(int(j))
print(discount_price)
```

```python
disc_product=soup.find_all("div",class_="product-discount")
discount_percnt=[]
for i in disc_product:
    j=i.text.replace("\n","").replace("%","").replace("Off","").strip()
    discount_percnt.append(int(j))
print(discount_percnt)
```

```python
ratting = []

for i in soup.find_all("div", class_="filled-stars"):
    style = i.get("style")
    if style:
        percent = style.split(":")[1].replace("%", "").strip()
        rating = round(float(percent) / 20,1)   # 100% = 5 stars
        ratting.append(rating)

print(ratting)
```

```python
review_count=soup.find_all("p",class_="product-rating-count")
review=[]
for i in review_count:
    j=i.text.replace("(","").replace(")","")
    review.append(int(j))
print(review)
```

```python
import pandas as pd
```

```python
max_len = max(
    len(brand),
    len(product_name),
    len(original_price),
    len(discount_price),
    len(discount_percnt)
)
max_len
```

```python
def fill_missing(lst, length, fill_value):
    return lst + [fill_value] * (length - len(lst))
```

```python
ratting = fill_missing(ratting, max_len, None)
review = fill_missing(review, max_len, 0)
```

```python
sr = pd.DataFrame({
    "brand": brand,
    "product_name": product_name,
    "original_price": original_price,
    "discount_price": discount_price,
    "discount_percent": discount_percnt,
    "ratting": ratting,
    "review": review
})
sr
```
